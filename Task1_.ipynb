```python
for column in years.columns:
    has_zero_or_negative = (years[column] <= 0).any()
    print(f"Column {column} has zero or negative numbers: {has_zero_or_negative}")
```

```python
upper_bounds = {}  # Dictionary to store upper bounds for each column

for column in years.columns:
    Q1 = years[column].quantile(0.25)
    Q3 = years[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    upper_bounds[column] = upper_bound  # Store the upper bound

for column, upper_bound in upper_bounds.items():
    print(f"Upper bound for column {column}: {upper_bound}")
```

```python
lower_bounds = {}  # Dictionary to store lower bounds for each column

for column in years.columns:
    Q1 = years[column].quantile(0.25)
    Q3 = years[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    lower_bounds[column] = lower_bound  # Store the lower bound

# Print the lower bounds that were used
for column, lower_bound in lower_bounds.items():
    print(f"Lower bound for column {column}: {lower_bound}")
```
